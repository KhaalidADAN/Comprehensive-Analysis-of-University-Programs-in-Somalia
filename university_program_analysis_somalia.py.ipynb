```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as nb
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = pd.read_excel("C:\\Users\\user\\Dropbox\\PC\\Downloads\\Merged Data.xlsx")
```

```python
# Perform descriptive statistics
descriptive_stats = data.describe(include='all')

# Filter relevant columns for further detailed statistics
relevant_stats = descriptive_stats[['Program Duration (Years)', 'Tuition Fees']]

# Display the relevant stats
print(relevant_stats)
```

```python
# Resetting the theme to default for best visualization without user-specific customization
sns.set(style="whitegrid")
plt.style.use("default")

# Re-plotting the distribution of program durations
plt.figure(figsize=(7, 5))
sns.histplot(data['Program Duration (Years)'].dropna(), bins=10, kde=True, color='skyblue')
plt.title('Distribution of Program Duration (Years)')
plt.xlabel('Program Duration (Years)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Re-plotting the distribution of tuition fees
plt.figure(figsize=(7, 5))
sns.histplot(data['Tuition Fees'].dropna(), bins=20, kde=True, color='skyblue')
plt.title('Distribution of Tuition Fees')
plt.xlabel('Tuition Fees')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot of Tuition Fees by Program Duration with default settings
plt.figure(figsize=(7, 5))
sns.boxplot(x='Program Duration (Years)', y='Tuition Fees', data=data, palette="coolwarm")
plt.title('Boxplot of Tuition Fees by Program Duration')
plt.xlabel('Program Duration (Years)')
plt.ylabel('Tuition Fees')
plt.show()
```

```python

# Scatter plot with regression line
plt.figure(figsize=(7, 5))
sns.scatterplot(x='Program Duration (Years)', y='Tuition Fees', data=data, color='blue')
sns.regplot(x='Program Duration (Years)', y='Tuition Fees', data=data, scatter=False, color='red')
plt.title('Correlation between Program Duration and Tuition Fees')
plt.xlabel('Program Duration (Years)')
plt.ylabel('Tuition Fees')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.violinplot(x='University Name', y='Tuition Fees', data=data, palette="coolwarm")
plt.title('Violin Plot of Tuition Fees Distribution by University')
plt.xlabel('University Name')
plt.ylabel('Tuition Fees')
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group by University and calculate the mean program duration
mean_duration = data.groupby('University Name')['Program Duration (Years)'].mean()

# Convert to DataFrame and reset index for heatmap
mean_duration_df = mean_duration.reset_index()

# Create a pivot table for heatmap
pivot_table = mean_duration_df.pivot_table(index='University Name', values='Program Duration (Years)')

# Plot the heatmap
plt.figure(figsize=(7, 5))
sns.heatmap(pivot_table, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title('Heatmap of Mean Program Duration Across Universities')
plt.ylabel('University Name')
plt.xlabel('Mean Program Duration (Years)')
plt.xticks(rotation=90)
plt.show()
```

```python
duration_fees = data.groupby('Program Duration (Years)')['Tuition Fees'].mean()
plt.figure(figsize=(7, 5))
plt.pie(duration_fees, labels=duration_fees.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("coolwarm", len(duration_fees)))
plt.title('Proportion of Tuition Fees by Program Duration')
plt.show()
```

```python
# Re-import necessary libraries and re-load the dataset to fix the issue

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset
file_path = ("C:\\Users\\user\\Dropbox\\PC\\Downloads\\Merged Data.xlsx")
data = pd.read_excel(file_path)

# Cleaning up and merging similar faculty names, assuming similar names represent the same faculty
# A dictionary for manual renaming (mapping similar faculty names to one suitable name)
faculty_rename_dict = {
    'Faculty of Economics & Management Sciences': 'Faculty of Economics and Management',
    'Faculty of Economics and Managements': 'Faculty of Economics and Management',
    'Faculty of Engineering, Information & Communications Technology': 'Faculty of Engineering and ICT',
    'Faculty of Engineering': 'Faculty of Engineering and ICT',
    'Faculty of Medicine & Health Sciences': 'Faculty of Medicine and Health Sciences',
    'Faculty of Medicine and Health Science': 'Faculty of Medicine and Health Sciences',
    'Faculty of Sharia & Law': 'Faculty of Sharia and Law',
    'Faculty of Computer Science & Information Technology': 'Faculty of Computer Science and IT',
    'Faculty of Computer Sciences & IT': 'Faculty of Computer Science and IT',
}

# Apply the renaming to the dataset
data['Faculty Name'] = data['Faculty Name'].replace(faculty_rename_dict)

# Recompute the count of programs by faculty after renaming
programs_by_faculty_merged = data.groupby('Faculty Name')['Program Name'].count()

# Sort the merged data in descending order
programs_by_faculty_sorted_merged = programs_by_faculty_merged.sort_values(ascending=False)

# Plotting the merged bar plot with different colors
plt.figure(figsize=(10, 8))
sns.barplot(y=programs_by_faculty_sorted_merged.index, x=programs_by_faculty_sorted_merged.values, palette='bright')

# Adding title and labels
plt.title('Number of Programs by Faculty (Merged and Cleaned)', fontsize=16)
plt.ylabel('Faculty Name', fontsize=12)
plt.xlabel('Number of Programs', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
